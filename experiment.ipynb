```python
#value
import numpy as np
random=np.array([0,23,250,691,721,385,164,48,20,5,2])
letter_freq =np.array([0,27, 386, 787,595,294,127,52, 27, 12,2])
entropy =np.array([0,40,530,1002,473,170,59,23,8,2,2])
entropy_wordfreq = np.array([0,52,706,1065,376,81,21,6,2])
Astar=np.array([0,110,724,947,362,123,31,8,4])
```

```python
import matplotlib.pyplot as plt

# Define the data to be plotted
x = ['Random', 'Letter Frequency', 'Entropy', 'Entropy+Word frequency', 'A*']
win_rate = [89.649, 90.472, 95.929, 98.744, 98.138]
average_score = [4.876, 4.673, 4.233, 3.924, 3.917]

# Find the peaks in the data
peaks = [89.649, 90.472, 95.929, 98.744, 98.138]

# Create a subplot
fig, ax = plt.subplots(figsize=(20, 10))

# Plot the data as a bar chart
ax.bar(x, average_score,color= '#c8b653', label='Average Score')
ax.set_ylim(3,5.5)

# Add labels and title to the subplot
ax.set_xlabel("Algorithms", fontsize =20)
ax.set_ylabel("Average Score",fontsize= 20)
ax.set_title("Average Score and Win Rate",fontsize=20)
for i in range(len(x)):
        plt.text(i, average_score[i], average_score[i], ha="center", va="bottom", size = 20)

# Add a legend
ax.legend(loc="upper left",fontsize=20)
ax2 = ax.twinx()
ax2.set_ylabel("Win Rate",fontsize=20)
#add line
ax2.plot(x, win_rate, label='Win Rate',color = '#6ca965', linewidth = 3)

# Add points at each peak
ax2.scatter(x,win_rate, color = '#6ca965',s = 100)
ax2.set_ylim(86,102)

for i in range(len(x)):
        if i==1:
                plt.text(i, win_rate[i] +1, win_rate[i], ha="center", va="bottom", size = 20)
        else:
                plt.text(i, win_rate[i] + 0.4, win_rate[i], ha="center", va="bottom", size = 20)

ax2.legend(fontsize=20)
# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

# Specify the data
labels = [f"{i}" for i in range(1,12)]
sizes = [i/sum(random) for i in random]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(20,20))

# Plot the pie chart
ax.pie(sizes, labels=labels, startangle=90, counterclock=False, autopct='%1.1f%%')

# Add a title to the plot
plt.title("Pie Chart Example")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Specify the data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 15, 25, 30]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the pie chart
ax.pie(sizes, labels=labels, startangle=90, counterclock=False, autopct='%1.1f%%')

# Add a title to the plot
plt.title("Pie Chart Example")

# Add labels to a box outside of the pie chart
text = ''
for i, label in enumerate(labels):
    text += f'{label}: {sizes[i]}%\n'

plt.text(1.5, 0.5, text, horizontalalignment='left',
         verticalalignment='center', transform=ax.transAxes)

# Show the plot
plt.show()
```
